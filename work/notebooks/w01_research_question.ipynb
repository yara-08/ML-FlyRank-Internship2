## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

Total pages: 30,000

Pages trending down: 16,262 (54.2%)

Average days since last update: 46.1 days

This dataset contains 30,000 content pages, providing a substantial amount of data for analysis. More than half of the pages (54.2%) are currently trending downward, indicating many potential refresh opportunities. On average, pages were last updated 46.1 days ago, suggesting that content freshness may be an important factor when deciding which pages should be prioritized for review.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.

import pandas as pd

#Load dataset
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

#statistics
total_pages = len(df)
down_pages = (df["trend_direction"] == "down").sum()
down_percentage = down_pages / total_pages * 100
avg_days_since_update = df["days_since_last_update"].mean()

print(f"Total pages: {total_pages:,}")
print(f"Pages trending down: {down_pages:,} ({down_percentage:.1f}%)")
print(f"Average days since last update: {avg_days_since_update:.1f} days")
```
